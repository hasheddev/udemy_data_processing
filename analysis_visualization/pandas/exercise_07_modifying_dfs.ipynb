```python
import pandas as pd

tweets = pd.read_csv('../data/JoeBidenTweets.csv')
tweets.set_index('id', inplace=True)
```

```python
tweets.drop(['url'], axis=1) #tweets.drop(columns='url')
```

```python
tweets.drop(361388562) #tweets.drop(index=361388562)
```

```python
tweets['user'] = 'Joe Biden'
tweets['ratio'] = tweets['replies'] / tweets['retweets']
```

```python
tweets.ratio.sort_values(ascending=False)
```

```python
tweets['interactions'] = tweets['replies'] +  tweets['retweets'] + tweets['quotes'] + tweets['likes']
```

```python
tweets['interactions'].sort_values(ascending=False)
```
