```python
import pandas as pd

titanic = pd.read_csv("../data/titanic.csv")

houses = pd.read_csv('../data/kc_house_data.csv')

netflix = pd.read_csv('../data/netflix_titles.csv', sep='|')
```

```python
houses[houses['price'] > 5000000]
```

```python
houses[houses['bedrooms'] >= 10]
```

```python
houses[houses['sqft_living'] < 500]
```

```python
#Between  filter in range
houses[houses['bedrooms'].between(5, 7)]
```

```python
houses[houses['grade'].between(11, 13)]['grade'].value_counts()
```

```python
#IS IN
netflix[netflix['country'].isin(['India', 'Japan', 'South Korea'])]
```

```python
netflix[netflix['rating'].isin(['TV-MA', 'R'])]
```

```python
is_female = titanic['sex'] == 'female'
died = titanic.survived == 0
titanic [is_female & died]
```

```python
waterfront = houses.waterfront == 1
price = houses.price < 500_000
houses[price & waterfront]
```

```python
houses[(houses.grade > 10) & (houses.view == 4)]
```

```python
houses[(houses.yr_built >= 2014) | (houses.yr_renovated >= 2014)]
```
